## String Prompttemplate

```python
from langchain_core.prompts import PromptTemplate
prompt_template=PromptTemplate.from_template('what is the definition of {word}')
prompt=prompt_template.invoke({word:'langchain'})
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model='gpt-4o-mini',temperature=0.9)
response=llm.invoke(prompt)
print(response.content)
```

## Chat prompttemplate

```python
from langchain_core.prompts import ChatPromptTemplate
chat_prompt_template=ChatPromptTemplate([('system','you are a helpful assistant'),
                                        ('human','what is the definition of {word}')])
chat_prompt=chat_prompt_template.invoke({'word':'langchain'})
print(chat_prompt)
```

```python
response=llm.invoke(chat_prompt)
print(response.content)
```

## Message PlaceHolder

```python
from langchain_core.prompts import MessagesPlaceholder,ChatPromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

message_prompt_template=ChatPromptTemplate.from_messages([SystemMessage(content='you are a helpful assistant'),
                                                        MessagesPlaceholder(variable_name='msg')])
## multiple human messages
messages= [HumanMessage(content='what is the definition of langchain'),
           AIMessage(content='langchain is a framework for building applications with language models'),
           HumanMessage(content='what are the features of langchain')]

#invoke the prompt template with the messages
output=message_prompt_template.invoke({'msg':messages})
print(output)
```

#### HumanMessage = what user said
#### AIMessage = what AI previously replied
#### MessagesPlaceholder = the slot that holds all of this history

#### You're manually building the memory. LangChain just structures it cleanly.

```python
response=llm.invoke(output)
print(response.content)
```
