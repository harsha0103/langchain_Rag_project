```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")
llm=ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
 # temperature is the creativity of the response, 0.9 is quite creative writing, 0.2 is more focused and deterministic, 
 # 0.0 is very deterministic and focused, and 1.0 is the most creative but also the most random.
 #  You can adjust this parameter based on your needs.
response = llm.invoke("Describe about langchain in one sentence")
response
```
