# Text based  Loader

```python
from langchain_community.document_loaders import TextLoader
path='text.txt'

loader = TextLoader(path)
docs= loader.load()
#print(len(docs))
print(docs[0].page_content)
#print(type(docs))
```

# PyPdf loader

```python
from langchain_community.document_loaders import PyPDFLoader

path="/Users/nirmalharshakolluru/Downloads/PDFs/hands_on_LLM .pdf"
pdf_loader= PyPDFLoader(path)
docs=pdf_loader.load()
print(docs[1].page_content)
```

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
path='/Users/nirmalharshakolluru/Downloads/PDFs/Textbooks'
pdf_loader = PyPDFDirectoryLoader(path,glob="**/*.pdf")
docs=pdf_loader.load()
print(f"Total pages: {len(docs)}")
print(f"Files loaded: {set(d.metadata['source'] for d in docs)}")
print(docs[0].page_content[:500])  # print first 500 chars only
```
